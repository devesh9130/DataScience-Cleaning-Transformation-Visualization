```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")  # Set a clean style for Seaborn
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/refs/heads/master/titanic.csv")
df
```

# Histogram

## A histogram is useful for showing the distribution of a variable like Age.

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title("Distribution of Passenger Ages")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

# Bar Charts

 ## A bar plot can compare counts of a categorical variable, like survival by Pclass.

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title("Survival Count by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.legend(title="Survived")
plt.show()
```

# Correlation

##  Calculate the correlation between numerical features in the dataset.


```python
correlation_matrix = df.select_dtypes(include='number').corr()
print(correlation_matrix) 
```

# Heatmaps

## Use Seabornâ€™s heatmap function to visualize the correlation matrix.

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Correlation Matrix")
plt.show()
```

# Subplots to Compare Multiple Variables
## Subplots allow you to show multiple charts side-by-side, making comparisons easier. 

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axs = plt.subplots(1, 2, figsize=(14, 6))

# Histogram of Age by Passenger Class
sns.histplot(df, x='Age', hue='Pclass', multiple='stack', ax=axs[0])
axs[0].set_title("Age Distribution by Passenger Class")
axs[0].set_xlabel("Age")
axs[0].set_ylabel("Frequency")

# Histogram of Fare by Passenger Class
sns.histplot(df, x='Fare', hue='Pclass', multiple='stack', ax=axs[1])
axs[1].set_title("Fare Distribution by Passenger Class")
axs[1].set_xlabel("Fare")
axs[1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```
