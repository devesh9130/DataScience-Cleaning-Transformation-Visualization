```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler, MinMaxScaler
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/refs/heads/master/titanic.csv")
df
```

# Label encoding for nominal variables.

## Label encoding is used for ordinal categorical variables, where the categories have a meaningful order or ranking. In our dataset, we might have variables that are ordered in a way that is relevant for analysis or modeling.


```python
label_encoder = LabelEncoder()
df['Gender'] = label_encoder.fit_transform(df['Sex'])
df
```

# One hot Encoding

## One-hot encoding is useful for categorical variables that have no inherent order or ranking among their values. 

```python
df = pd.get_dummies(df,columns=['Embarked'], drop_first=True)
df
```

# Feature Enginnering

## Feature engineering involves creating new features from the existing data that may better capture the underlying patterns. 

## Age Group

##  By creating age groups, you simplify age into categories, which might reveal patterns that raw ages could obscure

```python
df['AgeGroup'] = pd.cut(df['Age'], bins=[0, 12, 18, 35, 60, 100], labels=['Child', 'Teen', 'YoungAdult', 'Adult', 'Senior'])
df
```

# Aggregation Functions

## The groupby() function in Pandas allows you to group data based on one or more columns and then apply an aggregation function to each group.

## Mean

## We calculated the average age for each passenger class to see if age distributions differed among classes

```python
mean_age_by_class = df.groupby('Pclass')['Age'].mean()
print(mean_age_by_class)
```

## Sum

## Summing the fare for each passenger class provides insight into the total revenue generated per class.

```python
total_fare_by_class = df.groupby('Pclass')['Fare'].sum()
print(total_fare_by_class)
```
